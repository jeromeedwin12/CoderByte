**Separating numerical and categorical columns**

```python
# Pulling out names of numerical variables by conditioning dtypes NOT equal to object type

numerical_features = df.dtypes[df.dtypes != "object"].index
print("Number of Numerical features: ", len(numerical_features))
print()
print(numerical_features)
```

```python
# Pulling out names of categorical variables by conditioning dtypes equal to object type

categorical_features = df.dtypes[df.dtypes == "object"].index
print("Number of Categorical features: ", len(categorical_features))
print()
print(categorical_features)
```

**Plotting graphs to visualize the datasets**

```python
#plotting relations between numerical values with the world population percentage

f = pd.melt(df, id_vars=['World Population Percentage'],value_vars= numerical_features[numerical_features != 'World Population Percentage'])
g = sns.FacetGrid(f, col="variable",  col_wrap=5, sharex=False, sharey=False)
g = g.map(sns.distplot, "value", color="blue")
```

```python
#plotting a boxplot for the categorical values  

sns.set(font_scale=1)

f = pd.melt(df, id_vars=['World Population Percentage'], value_vars=categorical_features)
facetobject = sns.FacetGrid(f, col="variable", col_wrap=2, sharex=False, sharey=False, size=6)
facetobject.map(sns.boxplot, "value", "World Population Percentage", palette="Set3")
facetobject.fig.subplots_adjust(wspace=.25, hspace=0.25)


for ax in facetobject.axes.flat: 
    plt.setp(ax.get_xticklabels(), rotation=90)
```

```python
#boxplot between continent and world population percentage

sns.catplot(data=df, x="Continent", y="World Population Percentage", kind="box")
```

**Converting categorical to numerical**

```python
# Import label encoder
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()

# Encode labels in column 'CCA3'.
df['CCA3']= label_encoder.fit_transform(df['CCA3'])
df['CCA3'].unique()
```

```python
# Encode labels in column 'CCA3'.

df['Country']= label_encoder.fit_transform(df['Country'])
df['Country'].unique()
```

```python
# Encode labels in column 'CCA3'.

df['Capital']= label_encoder.fit_transform(df['Capital'])
df['Capital'].unique()
```

```python
# Encode labels in column 'CCA3'.

df['Continent']= label_encoder.fit_transform(df['Continent'])
df['Continent'].unique()
```
